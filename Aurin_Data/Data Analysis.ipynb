```python
import pandas as pd
import numpy as np
import json
```

```python
FILE_NAME = '/Users/gusyin/Documents/GitHub/COMP90024_Team29/harvester/mapreduce_result/bendigo.json'


f = open(FILE_NAME, 'r',encoding = "utf-8")
```

```python
bendigo_df = pd.DataFrame()
date = []
tweetcounts = []
approval_rate = []
with open(FILE_NAME) as json_file:
    data = json.load(json_file)
    for row in  data["daily_sentiment"]["rows"]:
        date.append(row['key'][0])
        tweetcounts.append(int(row['value'] / row['ratio']))
        approval_rate.append(row['ratio'])
bendigo_df['date'] = date
bendigo_df['tweetcounts'] = tweetcounts
bendigo_df['approval_rate'] = approval_rate

bendigo_df.head()
        
        
```

```python


import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(bendigo_df['date'],bendigo_df['approval_rate'])
plt.ylabel('Approval Rate')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 7))

color = 'tab:blue'
ax1.set_xlabel('Cities')
ax1.set_ylabel('tweet count', color=color)
ax1.bar(names, total_count_lst, color=color)
ax1.tick_params(axis='y', labelcolor=color)

def add_value_labels(ax, spacing=5):
    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.1f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.


# Call the function above. All the magic happens there.
add_value_labels(ax1)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis


color = 'tab:red'
ax2.set_ylabel('average approval rate', color=color)  # we already handled the x-label with ax1
ax2.plot(names, approval_rate_lst, color=color,linewidth=3,marker='D', markeredgecolor='black')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
